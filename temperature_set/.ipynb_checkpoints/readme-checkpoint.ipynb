### Problem statement
Provided a csv file each of four transfer curves of a set of nanowires at a range of temperatures, perform statistics and data analysis on the entire dataset. A versatile program that allows the analysis of a range of different aspects of the experiment. 

Each device has a device name and multiple trials are performed on the device, each to get four transfer curves (on and off at two potential difference values). 

> This project is a good opportunity to practice using some common libraries in Data Science (SciPy, Matplotlib, Pandas, Seaborn). 

# Work description
The preprocessing flow was 
* to parse each device's transfer curve data by opening and reading CSVs
* to split the transfer curve data into four segments (measurement device flattens the array)
* to repeat this process over a set of temperatures in multiple directories
* to store each set of data in a meaningful format with suitable headers

To analyse data in a versatile and customisable manner, we had to allow for variations in the groupings of what data to present in each graph, thus we had 
* to allow for the filtering of data
* to output transfer curve plots
* 
* to calculate the transconductance (derivative) of the curves, apply a polynomial fit (9th order) and find its maximum


There were a few iterations of this project, the first being solely based on SciPy, which was capable of analysing multiple devices at one ambient condition. The later iterations explored using JSONs to store and index the data, which was unsuccessful. The use of pandas triumphed because of its _versatility_ in the form of indexing and its ability to deal with _missing data_.

## Analysis of transfer curves
The analysis suite is capable of filtering what data to show in our plots and how many and what different plots to show. Note that our suite gives suitable titles and labels based on what is and is not common to all the data. 

### Comparison by temperature

#### Example 1 (Regular transfer curve plot)

This plot is standard and is shown on the measurement machine itself.

```python
### Setting adjustments for filters, stating which conditions/devices/trials/experiment to INCLUDE #####
temperatures = ['80K']
devices = ['0303d69']
trials = ['(1)']
experiments = ['all']

filters = [temperatures, devices, trials, experiments]
filtered = a.filterdf(a.df, filters)

### Setting the comparison order
order = ['temperatures', 'devices', 'trials', 'experiments']
split = 'temperatures'

plt.figure(figsize=(12,5))
a.genloop(order, split, filtered, 'self.plotTransfer(df, title)')
```

#### Example 2 Batch plot

We can plot the same plot of all the devices at once.

```python
### Setting adjustments for filters, stating which conditions/devices/trials/experiment to INCLUDE #####
temperatures = ['80K', '100K', '150K']
devices = ['0303d69']
trials = ['(1)']
experiments = ['all']

filters = [temperatures, devices, trials, experiments]
filtered = a.filterdf(a.df, filters)

### Setting the comparison order
order = ['temperatures', 'devices', 'trials', 'experiments']
split = 'trials'

plt.figure(figsize=(12,5))
a.genloop(order, split, filtered, 'plt.figure(figsize=(12,5)); self.plotTransfer(df, title)')
```

#### Example 3 (Combined plots of one device's transfer curves at multiple temperatures)

Good for comparison and gives a **visual indication** of each how the shape of the curves vary over temperatures.

```python
### Setting adjustments for filters, stating which conditions/devices/trials/experiment to INCLUDE #####
temperatures = ['all']
devices = ['0303d69']
trials = ['(1)']
experiments = ['10mV-off']

filters = [temperatures, devices, trials, experiments]
filtered = a.filterdf(a.df, filters)

### Setting the comparison order
order = ['temperatures', 'devices', 'trials', 'experiments']
split = 'temperatures'

plt.figure(figsize=(12,5))
a.genloop(order, split, filtered, 'plt.figure(figsize=(12,5)); self.plotTransfer(df, title)')
```

### Comparing device performances i.e. split plots by device

#### Example 4 (Comparing both trials of each device at the 10mV-off setting, both at 80K)

```python
### Setting adjustments for filters, stating which conditions/devices/trials/experiment to INCLUDE #####
temperatures = ['80K']
devices = ['0303d69', '0312a25']
trials = ['all']
experiments = ['10mV-off']

filters = [temperatures, devices, trials, experiments]
filtered = a.filterdf(a.df, filters)

### Setting the comparison order
order = ['devices', 'temperatures', 'trials', 'experiments']
split = 'devices'

plt.figure(figsize=(12,5))
a.genloop(order, split, filtered, 'plt.figure(figsize=(12,5)); self.plotTransfer(df, title)')
```

This shows that the for the same device, the second trial tends to have greater currents for the same ground-source voltage. It also shows that increased temperature leads to a decrease in current.
